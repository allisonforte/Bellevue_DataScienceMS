```python
# Load the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Read the dataset

df = pd.read_excel('/Users/allison.forte/Downloads/Weeks 5 & 6 Data/candyhierarchy2017.xlsx')


# Check the first 10 records

print(df.head(10))


# Find the total number of records

print('\nThe shape of this dataset is ', df.shape)
```

# Chapter 7 transformations

1- Filter out missing data

```python
#  assess the missing data

print(df.isnull().sum())


#  Remove the data that does not have an age so that analysis can be completed by age

df2 = df[df['Q3: AGE'].notna()]


#  Remove the data that does not have an answeer to 'going out' so analysis can be completed by going out or not and this can be compared to age

df2  = df2[df2['Q1: GOING OUT?'].notna()]


#  Confirm the appropriate columns no longer have missing data

print(df2.isnull().sum())


#  Assess the new shape of the data with missing data in key columns removed

print('\nThe new shape of this dataset is ', df2.shape)
```

# Chapter 7 transformations

2- Transform data using either mapping or a function

```python
#  Create a map to return a given value based on current value

dayToNum = {
            'sunday': 1,
            'monday': 2,
            'tuesday': 3,
            'wednesday': 4,
            'thursday': 5,
            'friday': 6,
            'saturday': 7,
            'none': 0}


#  Define the data that will be used in the function

days = (df2['Q11: DAY'].str.lower())  #  Ensure case will not be a problem
days = days.fillna('none')  # Define NA value to ensure it works with the function


#  Use map to add the new column to the data frame then print the result

df2['day_num_map'] = days.map(dayToNum)

df2.head()
```

# Chapter 8 transformation

1- Index with a column creating a hierarchical index

```python
#  Use 2 columns as indexes

df3 = df2.set_index(['Q1: GOING OUT?', 'Q2: GENDER'])


#  Summarize by one of the indexes

df3.groupby(level = 'Q1: GOING OUT?').count()
```

```python
#  Reshape the data by swapping the levels

df4 = df3.swaplevel('Q1: GOING OUT?','Q2: GENDER')


#  Summarize by one of the indexes

df4.groupby(level = 'Q2: GENDER').count()
```

# Chapter 10 transformation

1- Split/Apply/Combine

```python
#  Split data by whether the responder is going out and then by gender
#  Calculate mean age for each group and combine results back together 


df3['Q3: AGE'] = pd.to_numeric(df3['Q3: AGE'], errors='coerce')  #  First, change column to float

print(df3.dtypes)  #  Confirm correct data type to work with
 
df3['Q3: AGE'].groupby(['Q1: GOING OUT?', 'Q2: GENDER']).mean()  #  Split/Apply/Combine
```

# Chapter 10 transformation

2- Pivot table and Cross tab

```python
#  Use cross tab to compare going out, gender, and age. Look at size of each group using len


pd.crosstab(df2['Q1: GOING OUT?'], df2['Q2: GENDER'], df2['Q3: AGE'], aggfunc = len, margins = True)
```

```python
#  Use a pivot table to look at margins for different groups


df2['Q3: AGE'] = pd.to_numeric(df2['Q3: AGE'], errors='coerce')  #  Change column to float

df2.pivot_table(['Q3: AGE'], index = ['Q2: GENDER'], columns = 'Q1: GOING OUT?', margins = True)
```

# Chapter 11 transformation

1- Generate date range

```python
from datetime import datetime
from dateutil.parser import parse

#  Generate a date range from October 1 through November 15 with 2293 periods
#  these parameters were chosen to align with Halloween (subject of the survey) and the length of the dataframe

date_range = pd.date_range(start = '10/1/2010', end = '11/15/2010', periods = 2293)
date_range


#  Add date_range to dateframe as 'date' column

df2['date'] = date_range
df2  #  Check dataframe to comfirm column was added
```

# Chapter 11 transformation

2- Convert timestamps to periods and back

```python
#  Practice creating a period range
#  Not used on the dataframe
period_range = pd.period_range('10-1-2010', '11-15-2010', freq='W')
period_range
```

```python
#  Convert the date column to a weekly period and add to the dataframe in a new column (period)

df2['period'] = df2['date'].dt.to_period('W')

df2  #  Confrim new column by looking at the dataframe
```

```python
#  Convert period back to date and add to the dataframe in a new column (new date)
#  Note, specific days are lost iin this process, the final day of the period is used instead

df2['new_date'] = df2['period'].dt.to_timestamp(how = 'e')

df2  #  Confrim new column by looking at the dataframe
```

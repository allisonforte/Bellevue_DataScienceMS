```python
# combine the preceeding code and write a separate function that returns the whole number from the list generated from dropwhile 
# make sure that the return type is indeed a number not a string (this was confirmed with 'print(type(final_number)))'

import math


permutations(range(3))


def convert_to_num(stack):
    final_num = 0
    for i in range(0, len(stack)):
        final_num += (stack.pop() * (math.pow(10, i)))  # (10 to the power of 0=1, power of 1=10, 2=100)
    return final_num


for num_tuple in permutations(range(3)):
    num_stack = list(dropwhile(lambda x: x <= 0, num_tuple))
    print(convert_to_num(num_stack))
    
```

## Data Wrangling with Python: Activity 4 (pg. 59)

```python
#create a function to zip header, line, and fillvalue=none

from itertools import zip_longest
import os


def dict_from_csv(header, line):  # The function constructs a dict out of linee and header and fills key:values
    zip_line = zip_longest(header, line, fillvalue = None)  # missing values should result in None
    line_dict = {kv[0]:kv[1] for kv in zip_line}
    return line_dict
```

```python
# open the sales_record.cvs file by using r mode inside a with block

with open('/Users/allison.forte/Documents/GitHub/Data-Wrangling-with-Python/Lesson02/Activity04/sales_record.csv', 'r') as fd:
    first = fd.readline()  # read the first line
    print(first)  # check that the file is open
    header = first.replace("\n", "").split(",")  # use string methods to generate a list of all the column names
    print(header)
    for i, line in enumerate(fd):  # read the file line by line
        line_edited = line.replace("\n", "").split(",")
        d = dict_from_csv(header, line_edited)  # pass each line to a function with the list of headers
        print(d)  # print each entry as generated
        if i > 10:  # this can be run for as many iterations as necessary
            break
        
    
```

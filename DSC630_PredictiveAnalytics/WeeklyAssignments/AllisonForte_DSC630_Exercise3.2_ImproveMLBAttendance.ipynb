```python
# Load data

import pandas as pd
import numpy as np

df = pd.read_csv('/Users/allison.forte/Downloads/dodgers-2022.csv')
print(df.head())
df.describe()
```

## Findings and analysis

Baseball schedules depend on many factors and require games against all teams. The opponent, while it can affect the attendance, is not something that can be easily changed. Factors including time of day and day of week can more easily be changed but are still often determined by more factors than just management decision. The easiest factors for management to change are whether games have free give aways or additional perks. This dataset has information about cap, shirt, and bobblehead giveaways as well as fireworks.

The max attendance in this dataset was 56,000 people and the minimum attendance was 24,312 people with all the data provided. The mean attendance was 41040 people. Tuesdays are the most attended day of the week with an average attendance of 47741.

```python
# calculate the average attendance for day games vs night games

table1 = pd.pivot_table(df, values=['attend'], index=['day_night'],
                    aggfunc={'attend': np.mean})

table1
```

With the current schedule and attendance, we can see that the time of day alone does not make a significant difference in the overall attendance of the game. Adjusting the time of games will likely not improve attendance in a significant way. 

```python
# calculate attendance by day of week

table1 = pd.pivot_table(df, values=['attend'], index=['day_of_week'],
                    aggfunc={'attend': np.mean}, sort = False)

table1
```

Additionally, looking the attendance by day of the week, it is clear that Monday games have the lowest attendance with Wednesday games being a close second. To improve attendance, fewer games should be scheduled on Mondays and Wednesdays with a preferred game day of Tuesday followed by Saturday then Sunday. With this information, it is important to remember that adjusting the day of a game is not as simple as a quick switch. The opponent's schedule as well as other factors must be considered before this recomendation can be enacted. 

Looking at thefactors that are more easily controlled, caps, shirts, fireworks, and bobbleheads all have different impacts on attendance. 

```python
# Check the overall impact of caps

cap = pd.pivot_table(df, values=['attend'], index=['cap'],
                    aggfunc={'attend': np.mean})                 
cap
```

```python
# Check the impact of caps by day of the week

cap2 = pd.pivot_table(df, values=['attend'], index=['cap', 'day_of_week'],
                    aggfunc={'attend': np.mean})
cap2
```

Based on the data provided by this dataset, caps should no longer be given out. Overall caps bring down attendance. Caps on Tuesdays significantly reduce attendance. Caps on Sunday lead to a very small increase in attendance. The expense spent on caps should be diverted to a more productive incentive.

```python
# Check the overall impact of shirts

shirt = pd.pivot_table(df, values=['attend'], index=['shirt'],
                    aggfunc={'attend': np.mean})
shirt
```

```python
# Check the impact of shirts by day of the week
                    
shirt2 = pd.pivot_table(df, values=['attend'], index=['day_of_week','shirt'],
                    aggfunc={'attend': np.mean})
shirt2
```

Shirts, unlike caps, have an overall positive impact on attendance with an average attendance on shirt nights being 46643 vs 40824 without shirts. Shirts are a significant draw on Sunday and Monday but have an opposite affect on Tuesdays. Shirts, like caps, should not be given out on Tuesdays. 

```python
# Check the overall impact of fireworks

fireworks = pd.pivot_table(df, values=['attend'], index=['fireworks'],
                    aggfunc={'attend': np.mean})
fireworks
```

```python
# Check the impact of fireworks by day of the week
                
fireworks2 = pd.pivot_table(df, values=['attend'], index=['day_of_week','fireworks'],
                    aggfunc={'attend': np.mean})
fireworks2
```

Overall, fireworks do not have a significant impact on attendance bringing the average attendance from 41032 to 41078. However, the fireworks on Wendesdays bring significantly more people increasing the average Wednesday night attendance from 36132 to 53570. There is not enough data to conclude whether frieworks on Fridays increase attendance. Based on this information, I would recomend moving half of the Friday fireworks shows to Wednesday or Monday which would have 2 positive effects. First, it will likely increase attendance on those days and second, it would allow the impact of fireworks on Friday to be further studied. 

```python
# Check the overall impact of bobbleheads

bobblehead = pd.pivot_table(df, values=['attend'], index=['bobblehead'],
                    aggfunc={'attend': np.mean})
bobblehead
```

```python
# Check the impact of bobbleheads by day of the week

bobblehead2 = pd.pivot_table(df, values=['attend'], index=['day_of_week','bobblehead'],
                    aggfunc={'attend': np.mean})
bobblehead2
```

Bobbleheads have the most significant impact on attendance overall. The average with bobblehead giveaways increases from 39138 to 53145. Given Monday has the lowest attendance of the whole week, adding some bobblehead days on Mondays will likely show a positive impact. Wednesdays would also likely benefit from bobbleheads.

Overall, based on what has been seen with this dataset, I would make several recomendations for improving attendance. First, whenever possible, avoid scheduling Monday and Wednesday games. In addition, ensure there is a game every Tuesday to maximize the most attended day of the week.
